```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# import weather data
weather = pd.read_csv("../data/weather/weather_data.csv")
```

```python
# TODO: Merge these two dataframes together
db = pd.merge(coffee, weather, on=['year', 'subdivision'])
print(db)
```

```python
# TODO: View the first 5 rows of your data
print(db.head)
```

```python
# TODO: Check for null values 

db.isnull()
```

```python
# TODO: Drop rows that contain null values

total = db.dropna()
```

```python
# TODO: BEGIN YOUR EDA!
#What is the relationship between temp_avg and 60kgs_bag? How strong is this relationship? Use pearson's correlation coefficient.
total.plot(
    x="60kgs_bag",              
    y="temp_avg",              
    kind='barh',    
    title="Temperature x coffee bag yield", 
    figsize=(10, 6),     
    xlabel='Coffee bags',         
    ylabel= 'temperature')
bagtempcorr = total["60kgs_bag"].corr(total["temp_avg"])
print(f"Pearon's correlation coefficient for temp_avg and 60kgs_bags is {bagtempcorr}, so the relationship is almost neutral strength")
    #What is the relationship between minimum humidity and 60kgs_bag? How *strong* is this relationship? Use pearson's correlation coefficient.
total.plot(
    x="60kgs_bag",              
    y="hum_min",              
    kind='barh',    
    title="Temperature x coffee bag yield", 
    figsize=(10, 6),     
    xlabel='Coffee bags',         
    ylabel= 'humidity')
baghumcorr = total["60kgs_bag"].corr(total["hum_min"])
print(f"Pearon's correlation coefficient for minimum humidity and 60kgs_bag is {baghumcorr}, so the relationship is weak")
#What is the relationship between rain_max and 60kgs_bag? How strong is this relationship? Use pearson's correlation coefficient.
total.plot(
    x="60kgs_bag",
    y="rain_max",
    kind='barh',
    title="Rain x coffee bag yield",
    figsize=(10, 6),
    xlabel='Coffee bags',
    ylabel='max rain levels'
           )
bagraincorr = total["60kgs_bag"].corr(total["rain_max"])
print(f"Pearon's correlation coefficient for rain_max and 60kgs_bag is {bagraincorr}, so the relationship is weak")
#What is the relationship between average wind speed and 60kgs_bag? How strong is this relationship? Use pearson's correlation coefficient.
total.plot(
    x="60kgs_bag",
    y="wind_avg",
    kind='barh',
    figsize=(10, 6),
    xlabel='Coffee bags',
    ylabel='Average wind speed',
)
bagwindcorr = total["60kgs_bag"].corr(total["wind_avg"])
print(f"Pearon's correlation coefficient for rain_max and 60kgs_bag is {bagwindcorr}, so the relationship is weak")
#Perform the same bivariate analysis between the mentioned weather columns (temp_avg, hum_min, rain_max, and wind_avg) and a coffee-related column of your choice (bearing_trees, nonbearing_trees, bearing_trees_per_hectare, etc). Which directions do these relationships follow? What are the strengths of these relationships?
total.plot(
    x="bearing_trees",
    y="hum_min",
    kind='barh',
    title="Humidty effects on fruit bearing trees",
    figsize=(10, 6),
    xlabel='bearing trees',
    ylabel='minimum humidity',
)
btreehumcorr = total["bearing_trees"].corr(total["hum_min"])
print(f"Pearon's correlation coefficient for Humidty effects on fruit bearing trees is {btreehumcorr}, so the relationship is weak")
total.plot(
    x="nonbearing_trees",
    y="hum_min",
    kind='barh',
    title="Humidity effects on nonfruit bearing trees",
    figsize=(10, 6),
    xlabel='nonbearing trees',
    ylabel='minimum humidity',
)
nbtreehumcorr = total["nonbearing_trees"].corr(total["hum_min"])
print(f"Pearon's correlation coefficient for Humidity effects on nonfruit bearing trees is {nbtreehumcorr}, so the relationship is below average")
total.plot(
    x="bearing_trees_per_hectare",
    y="wind_avg",
    kind='barh',
    title="Wind's effect on bearing trees per hectare",
    figsize=(10, 6),
    xlabel='bearing trees per hectare',
    ylabel='average wind speed',
)
btreehwindcorr = total["bearing_trees_per_hectare"].corr(total["wind_avg"])
print(f"Pearon's correlation coefficient for Wind's effect on bearing trees per hectare is {btreehwindcorr}, so the relationship is weak")
total.plot(
    x="hum_min",
    y="wind_avg",
    kind='barh',
    title="wind effect on minimum humidity",
    figsize=(10, 6),
    xlabel='Minimum Humidity',
    ylabel='Average wind speed',
)
hmwindcorr = total["hum_min"].corr(total["wind_avg"])
print(f"Pearon's correlation coefficient for wind effect on minimum humidity is {hmwindcorr}, so the relationship is averagely weak")
#Which weather factor seems to influence coffee production the most?
#ANSWER:Humidity
#Which data would you collect to continue exploring the impacts of weather and economic forces on coffee production?
#ANSWER:HUMIDITY, RAIN FALL, FRUIT BEARING PER HECTARE, 60KGS BAGS
#If you were preparing a report to provide coffee farmers, which weather condititions would you state are the most favorable for coffee growing (assuming any exist)?

```
