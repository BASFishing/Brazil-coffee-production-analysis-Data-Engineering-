```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather = pd.concat([weather1, weather2])
weather.to_csv("weather_data.csv")
```

```python
# TODO: view first 5 rows of your data
# WHy isn't 5th row things
weather.head()
```

```python
# TODO: describe your numeric columns
weather.describe()
```

```python
# TODO: BEGIN YOUR EDA!(Exploratory data analysis)
#Univariate Analysis

#How is average temperature, minimum humidity, and average wind speed changing across the years for Minas Gerais? Is this increasing or decreasing?
#ANSWER: EACH CATEGORY SEEMS TO BE TRENDING DOWNWARDS BUT RAIN, HUMIDITY AND TEMPERATURE SEEM TO SHIFT EVER 3 YEARS.
#Which year had the highest maximum record rainfall?
#ANSWER: 2006 HAD THE HIGHEST RAINFALL

weather.plot(
    x="year",              
    y="temp_avg",           
    kind='line',      
    title="Average Temperature of Minas Gerais",
    figsize=(10, 6),    
    xlabel='year',       
    ylabel='temperature',   
    )
weather.plot(
    x="year",        
    y="hum_min",             
    kind='line',         
    title="Humidity lows in Minas Gerais", 
    figsize=(10, 6),    
    xlabel='year',     
    ylabel='humidity',   
    )
weather.plot(
    x="year",            
    y="wind_avg",            
    kind='line',       
    title="Average wind speeds in Minas Gerais",
    figsize=(10, 6),     
    xlabel='year',        
    ylabel='wind speed',   
    )
weather.plot(
    x="year",           
    y="rain_max",            
    kind='line',        
    title="Rain in Minas Gerais", 
    figsize=(10, 6),  
    xlabel='year',         
    ylabel='rain fall',   
    )

#Bivariate Analysis
#What is the relationship between max rain and max humidity? How strong is this relationship? Use pearson's correlation coefficient.
#ANSWER:hIGH HUMIDITY HIGH RAIN
#What is the relationship between average wind speed and average temperature? How strong is this relationship? Use pearson's correlation coefficient.
#year,rain_max,temp_avg,temp_max,temp_min,hum_max,hum_min,wind_max,wind_avg,subdivision

weather.plot(
    x="rain_max",              
    y="hum_max",              
    kind='scatter',    
    title="Rain & Hummdity in Minas Gerais", 
    figsize=(10, 6),     
    xlabel='Rain levels',         
    ylabel='Humidity levels',   
    )
RH_correlation = weather['rain_max'].corr(weather['hum_max'])
weather.plot(
    x="wind_avg",             
    y="temp_avg",             
    kind='scatter',         
    title="Wind & temperature in Minas Gerais", 
    figsize=(10, 6),    
    xlabel='wind speed',      
    ylabel='temperature',   
    )
WT_correlation = weather['wind_avg'].corr(weather['temp_avg'])

print(f"Pearson's correlation coefficient for Rain and Humudity is: {RH_correlation}, indicating a positive correlation")

print(f"Pearson's correlation coefficient for Wind speed and temperature is: {WT_correlation}, indicating a positive correlation")
```
