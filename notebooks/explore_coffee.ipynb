```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isnull().sum()
coffee.isnull()
```

```python
# TODO: drop your null values
coffee_clean = coffee.dropna()
```

```python
# TODO: Select only Minas Gerais

MCoffee = coffee_clean[coffee_clean['subdivision'] == 'Minas Gerais']
```

```python
# TODO: BEGIN YOUR EDA!
#What is the relationship between bearing_trees and 60kgs_bag in Minas Gerais? How *strong* is this relationship? Use pearson's correlation coefficient.



sns.lineplot(MCoffee, x="year", y="bearing_trees")
plt.show()

coffee_correlation = MCoffee["bearing_trees"].corr(MCoffee["60kgs_bag"])
print(f"Pearon's correlation coefficient is {coffee_correlation}, so the relationship is strong")

#What is the relationship between bearing_hectares and 60kgs_bag in Minas Gerais? How *strong* is this relationship? Use pearson's correlation coefficient.

sns.lineplot(MCoffee, x="year", y="bearing_hectares")
plt.show()

coffee_correlation2 = MCoffee["bearing_hectares"].corr(MCoffee["60kgs_bag"])
print(f"Pearon's correlation coefficient is {coffee_correlation2}, so the relationship is weak")

sns.lineplot(MCoffee, x="year", y="60kgs_bag")
plt.show()
coffee_correlation3 = MCoffee["year"].corr(MCoffee["60kgs_bag"])
print(f"Pearon's correlation coefficient is {coffee_correlation3}, so the relationship is  strong")

sns.lineplot(MCoffee, x="60kgs_bag", y="bearing_hectares")
plt.show()
coffee_correlation4 = MCoffee["bearing_hectares"].corr(MCoffee["60kgs_bag"])
print(f"Pearon's correlation coefficient is {coffee_correlation4}, so the relationship is weak")

```
